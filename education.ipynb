# 4. Prediction Model

```python
import pandas as pd
df = pd.read_csv('enhanced_student_habits_performance_dataset.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sb
```

```python
cat_cols = df.select_dtypes(include='object').columns # columns indecies
```

```python
print(cat_cols) # index of columns which are object types
```

```python
df_encoded = pd.get_dummies(df, columns=cat_cols,dtype=int,drop_first=True)
```

```python
df_encoded.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

X = df_encoded.drop('exam_score', axis=1)
y = df_encoded['exam_score']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestRegressor()
model.fit(X_train, y_train)

print("Test Accuracy:", model.score(X_test, y_test))
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

corr = df_encoded.corr()

plt.figure(figsize=(30,30))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix (Numeric columns only)')
plt.show()
```

```python
corr = df.select_dtypes(include=['float','int']).corr().abs()
upper = corr.where(~np.tril(np.ones(corr.shape)).astype(bool))
top_corr = upper.stack().sort_values(ascending=False).head(15)
```
