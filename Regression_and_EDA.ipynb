문제 4-1. 이슬점 온도 계산 공식 구하기 이슬점 온도 구하는 식을 Regression 기법을 통해 구해봅시다. 이슬점 온도 구하는 식은 미지수 a, b, c, d, e에 대해서 아래와 같다고 가정합니다. 이슬점 온도 = a * 습도 + b * 풍속 + c * 현지기압 + d * 기온 + e 이 때 a, b, c, d, e 값을 구하여 아래에 답하세요.


```python
# 변수가 네개 : 습도, 풍속, 현지기압, 기온 -- 각각 x1, x2, x3, x4 네개의 입력 변수로 간주, 예측하고자 하는 값은 y로 이슬점 온도이다.
```

```python
import torch

x_train = torch.FloatTensor([df['습도'], df['풍속'], df['현지기압'], df['기온']])
y_train = torch.FloatTensor([df['이슬점 온도']])
```

```python
x_train
```

```python
y_train
```

```python
x_train = x_train.t()
x_train.shape
```

```python
y_train = y_train.t()
y_train.shape
```

```python
x_train.shape, y_train.shape
```

```python
W = torch.rand(4,1) # 4x1
b = torch.rand(1,1) # 1x1

lr = 0.01
print(W,b)
```

```python
lr = 0.0001

for epoch in range(3001): # How many times gonna repeat
  W.requires_grad_(True)
  b.requires_grad_(True) # calculating Gradient  // requires grad True

  # making hypothesis
  h = torch.mm(x_train, W) + b

  # gap btw true values vs hypothesis values
  cost = ((y_train - h) ** 2).mean()

  cost.backward() # Gradient calcualating function
  with torch.no_grad() as grd : # pause _ gradient calculating
    W = W - lr * W.grad # updating W value : moving - direction
    b = b - lr * b.grad

  if epoch % 100 == 0: # checking the learning process
    print(epoch, cost.item(), W.squeeze(), b)
```

```python
import pickle
data = pickle.load(open("mid_animal_data_pub.pkl", "rb"))
```

```python
data.keys()
```

```python
# train_vectors : train images의 사진들을 256차원의 벡터로 표현한 numpy array / 두 사진의 동물이 같으면 cosine 유사도가 높음
# train_labels : 0:고양이 / 1:강아지 / 2:다람쥐
# test1_images : 10장의 사진으로 생성한 numpy array
# test1_vectors : test1_images와 상응하는 벡터로 포함 , numpy array
# test2_vectors : 랜덤(익명)의 이미지 30장을 벡터로 표현
```

```python
import matplotlib.pyplot as plt

for i in range(10):
  plt.subplot(2, 5, i+1)
  plt.imshow(data['test1_images'][i])
  plt.axis('off')
```

```python
import numpy as np
from scipy.spatial.distance import cosine

1 - cosine(data['train_vectors'][0], data['test1_vectors'][1])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
test_vectors = data['test1_vectors']
train_vectors = data['train_vectors']

similarities =[]

for i in test_vectors :
  cosine_sim = cosine_similarity([i], train_vectors)[0]

  top_index = np.argsort(cosine_sim)[-3:][::-1]
  similarities.extend(top_index)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

x_train = torch.from_numpy(data["train_vectors"])
y_train = torch.from_numpy(data["train_labels"])

model = nn.Linear(256,3)
optim = torch.optim.Adam(model.parameters(),lr=0.01)

for epoch in range(10001):
  z = model(x_train)
  cost = F.cross_entropy(z,y_train)

  optim.zero_grad()
  cost.backward()
  optim.step()

  with torch.no_grad():
    if epoch % 1000 == 0:
       print(f"epoch: {epoch}, cost: {cost.item()}")
```

```python
data['test2_vectors']
```

```python
new_input = torch.from_numpy(data['test2_vectors'])

with torch.no_grad():
    predictions = model(new_input)

predictions
```

```python
# Predicting
result_string = ""

for prediction in predictions:
    predicted_class = torch.argmax(prediction).item()
    result_string += str(predicted_class)

print(result_string)
```
