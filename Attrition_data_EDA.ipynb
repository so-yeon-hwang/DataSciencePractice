```python
import pandas as pd
df = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition_2.csv')
```

```python
df
```

```python
df.columns
```

```python
df['Age'].value_counts()
```

```python
# Age column의 값을 나이대 별로 분류 (10s, 20s, 30s, 40s, 50s,60s)하여 탐색하기
# 나이 대 별 column 생성 - 해당 나이대의 값들만 작성되고, 나머지는 NaN값으로 들어감

filt = df['Age'] // 10 == 1
df['teen'] = df['Age'][filt]
df['teen']

filt = df['Age'] // 20 == 1
df['twenty'] = df['Age'][filt]
df['twenty']

filt = df['Age'] // 30 == 1
df['thirty'] = df['Age'][filt]
df['thirty']

filt = df['Age'] // 40 == 1
df['fourty'] = df['Age'][filt]
df['fourty']

filt = df['Age'] // 50 == 1
df['fifty'] = df['Age'][filt]
df['fifty']

filt = df['Age'] // 60 == 1
df['sixty'] = df['Age'][filt]
df['sixty']
```

```python
# 먼저 만든 10대-60대 column을 하나로 합쳐준다.
# 그리고 만들어둔 나이대 별 column은 조건 성립 여부를 따지는 도구로만 활용하고 나중에 드롭시킬 것.

# 컬럼 합치기
# concat 활용
ages = [ df['teen'], df['twenty'], df['thirty'], df['fourty'], df['fifty'], df['sixty'] ]
age_range = pd.concat(ages,axis=1) # new DataFrame 생성
age_range

```

```python
age_range
```

```python
col_name = list(age_range.columns)
col_name
```

```python
df['Ages'] = ''
df['Ages']
```

```python
col_dic = {
    'teen': 10,
    'twenty': 20,
    'thirty': 30,
    'fourty' : 40,
    'fifty' : 50,
    'sixty' : 60
}
```

```python
for col, age in col_dic.items():
    df.loc[df[col].notnull(), 'Ages'] = age
```
