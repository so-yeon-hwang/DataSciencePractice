```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

inertia = []

# 시도할 클러스터 수의 범위 설정 (예: 1에서 10까지 시도)
k_values = range(1, 11)

# Elbow Method 수행
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data)
    inertia.append(kmeans.inertia_)

# Elbow Method 그래프 그리기
plt.plot(k_values, inertia, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('The Elbow Method for Optimal k')
plt.show()
```

```python
import numpy as np
from sklearn.cluster import KMeans

# 최적의 클러스터 수
optimal_k = 8

# KMeans 모델 초기화 및 학습
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
kmeans.fit(data)

# 각 데이터 포인트에 대한 클러스터 할당 얻기
cluster_assignments = kmeans.predict(data)

# 결과 출력
for i, cluster in enumerate(cluster_assignments):
    print(f"Data point {i} is assigned to cluster {cluster}")
```

```python
import numpy as np

# 주어진 드라마 평점 행렬
ratings = np.array([
    [2, 3, 5, 2, 3],
    [2, 4, np.nan, 2, 4],
    [5, np.nan, 4, 2, 3],
    [5,4,1,3,2],
    [2,4,3,5,1]
])

# 중심화된 코사인 유사도 계산 함수
def centered_cosine_similarity(a, b):
    a_mean = np.nanmean(a)
    b_mean = np.nanmean(b)
    centered_a = a - a_mean
    centered_b = b - b_mean
    numerator = np.nansum(centered_a * centered_b)
    denominator = np.sqrt(np.nansum(centered_a**2) * np.nansum(centered_b**2))
    similarity = numerator / denominator if denominator != 0 else 0
    return similarity

# 드라마 간의 유사도 행렬 계산
num_dramas = ratings.shape[1]
similarity_matrix = np.zeros((num_dramas, num_dramas))

for i in range(num_dramas):
    for j in range(i, num_dramas):
        if i != j:
            similarity_matrix[i, j] = centered_cosine_similarity(ratings[:, i], ratings[:, j])
            similarity_matrix[j, i] = similarity_matrix[i, j]

# 시청자 C가 시청한 드라마 'SKY캐슬'에 대한 예상 별점 계산
target_drama_index = 1  # 'SKY캐슬'에 해당하는 열의 인덱스
target_drama_ratings = ratings[:, target_drama_index]

# 가장 유사한 2개 드라마 찾기
most_similar_dramas = np.argsort(similarity_matrix[target_drama_index, :])[-3:-1]

# 유사도를 가중치로 사용하여 예상 별점 계산
weighted_sum = 0
weight_sum = 0
for similar_drama in most_similar_dramas:
    if not np.isnan(target_drama_ratings[similar_drama]):
        weighted_sum += similarity_matrix[target_drama_index, similar_drama] * target_drama_ratings[similar_drama]
        weight_sum += np.abs(similarity_matrix[target_drama_index, similar_drama])

predicted_rating = weighted_sum / weight_sum if weight_sum != 0 else np.nan

print(f"예상 별점(SKY캐슬): {predicted_rating:.3f}")
```

```python
import numpy as np

def func(vec1, vec2):
    # Subtract the mean from each vector
    vec1_centered = vec1 - np.mean(vec1)
    vec2_centered = vec2 - np.mean(vec2)

    # Calculate the dot product
    dot_product = np.dot(vec1_centered, vec2_centered)

    # Calculate the norms
    norm_vec1 = np.linalg.norm(vec1_centered)
    norm_vec2 = np.linalg.norm(vec2_centered)

    # Calculate centered cosine similarity
    similarity = dot_product / (norm_vec1 * norm_vec2)

    return similarity

# Example usage with the provided vectors
vector1 = np.array([-1., 0., 2., -1., 0.])
vector2 = np.array([2., 1., -2., 0., -1.])

result_similarity = func(a, b)
print("Centered Cosine Similarity:", result_similarity)
```
