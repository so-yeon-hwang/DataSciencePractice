```python
import tensorflow as tf
import pandas as pd
import numpy as np
```

# 1. Data Import

```python
df = pd.read_csv('gpascore.csv')
df.head()
```

# 2. Data PreProcessing

```python
df.isnull().sum()
```

```python
df = df.dropna() # or data.fillna(avg)
```

```python
df.isnull().sum()
```

```python
y_train = df['admit'].values
y_train
```

```python
x_train = []
for i,rows in df.iterrows():
  x_train.append([rows['gre'], rows['gpa'],rows['rank']])
```

```python
x_train
```

```python
x_train = np.array(x_train)
```

```python
y_train = np.array(y_train)
```

# 3. Modeling

```python
# Building Deep Learning Model
# keras : tool for making deep learning model

model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(64, activation = 'tanh'),
    tf.keras.layers.Dense(128, activation = 'tanh'),
    tf.keras.layers.Dense(1, activation = 'sigmoid')
])

# model compiling
# optimizer : moderating learning rate

model.compile(optimizer = 'adam' , loss = 'binary_crossentropy', metrics = ['accuracy'])

# model fit
model.fit(x_train, y_train, epochs = 1000)
```

# 4. Predicting New data

```python
model.predict([[750, 3.70, 3], [400, 2.2, 1]]) # 0.69, 0.02
```
