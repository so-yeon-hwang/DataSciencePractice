```python
x = df['전월 대비 증감수']
y = df['전월 대비 증감율']

plt.scatter(x, y)

y_mid = 0.125
x_mid = 100000

plt.axhline(y=y_mid, color='black',alpha = 0.3, linestyle = ':',label='Horizontal Line')
plt.axvline(x=x_mid, color='black',alpha = 0.3,  linestyle = ':',label='Vertical Line')

x_range = [20000, 60000, 100000, 140000, 180000]
plt.xticks(x_range)
```
