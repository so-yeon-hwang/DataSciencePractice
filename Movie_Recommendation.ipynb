```python
# genre : from json string to python set
# make a new function

def str_to_set(x):
  genre_set = set() # make an empty set
  for item in eval(x): # eval function > python object
    genre_set.add(item['name'])
  return genre_set
```

```python
# apply the function

meta.genres = meta.genres.apply(str_to_set)
meta
```

# 4. Jaccard similarity

```python
# comparing two movies : The Dark Knight & The Dark Kngiht Rises
```

```python
dk = meta.loc[meta.title == 'The Dark Knight'].iloc[0] # only located in zero row
dkr = meta.loc[meta.title == 'The Dark Knight Rises'].iloc[0]
pd.concat([dk,dkr],axis=1).T
```

```python
# Jaccard similarity Function

def jaccard_sim(s1,s2):
  if len(s1|s2) == 0: # if the size is zero, cannot calculate denominator
    return 0
  return len(s1&s2)/len(s1|s2) # size of the group

jaccard_sim(dk.genres|dk.keywords,dkr.genres|dkr.keywords)
```

# 6. Pearson Correlation Coefficient

```python
def pearson_sim(m1,m2):
  m1_c = m1 - m1.mean()
  m2_c = m2 - m2.mean()

  denom = np.sqrt(np.sum(m1_c **2) * np.sum(m2_c ** 2))

  if denom != 0 :
    return np.sum(m1_c * m2_c)/denom
  else:
    return 0
```

```python
dk_rating = matrix['The Dark Knight']
pk_rating = matrix['Prom Night']
pearson_sim(dk_rating,pk_rating)
```

# 7. Recommend Similar Movie Function

```python
matrix
```

```python
def find_similar_movies (input_title , matrix, n, alpha):
 input_meta = meta.loc[ meta[ 'title'] == input_title].iloc[ 0]
 input_set = input_meta.genres | input_meta.keywords

 result = []

 for this_title in matrix.columns:
    if this_title == input_title:
      continue

    this_meta = meta.loc[meta['title'] == this_title].iloc[0]
    this_set = this_meta.genres | this_meta.keywords

    pearson = pearson_sim(matrix[this_title], matrix[input_title])
    jaccard = jaccard_sim(this_set, input_set)

    score = alpha * pearson + (1-alpha) * jaccard
    result.append((this_title, pearson, jaccard, score))

 result.sort(key= lambda r: r[3], reverse= True)

 return result[:n]
```

```python
result = find_similar_movies('The Dark Knight', matrix, 10, 0.3)
pd.DataFrame(result, columns = ['title', 'pearson', 'jaccard', 'score'])
```
