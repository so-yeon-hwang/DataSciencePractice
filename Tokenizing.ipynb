```python
import nltk
from nltk.tokenize import word_tokenize

# NLTK의 다운로드 매니저를 사용하여 필요한 자원 다운로드
nltk.download('punkt')

# 텍스트 데이터
text = "NLTK is a leading platform for building Python programs to work with human language data."

# 단어 토크나이저를 사용하여 텍스트를 단어 단위로 토크나이즈
tokens = word_tokenize(text)

# 결과 출력
print("토큰화된 단어들:", tokens)
```

```python
# 데이터프레임의 특정 열에 있는 텍스트 데이터를 토크나이즈하는 함수
def tokenize_dataframe_column(df, column_name):
    tokenized_texts = []
    for text in df[column_name]:
        tokens = word_tokenize(text)  # 텍스트를 단어 단위로 토크나이즈
        tokenized_texts.append(tokens)
    return tokenized_texts
```

```python
tokenized_texts = tokenize_dataframe_column(df, 'text')
```

```python
tokenized_texts
```

```python
from collections import Counter

# 모든 형태소를 하나의 리스트로 펼치기
all_tokens = [token for text_tokens in tokenized_texts for token in text_tokens]

# 형태소의 등장 빈도 계산
word_counts = Counter(all_tokens)

# 가장 많이 등장하는 단어 5개 추출
top_keywords = word_counts.most_common(120)

# 결과 출력
print("가장 많이 등장하는 단어 15개:")
for keyword, count in top_keywords:
    print(keyword, ":", count)

# California, disaster, buildings, Youtube
```
