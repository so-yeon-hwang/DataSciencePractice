```python
category_counts = df['category'].value_counts()
# visualization
plt.figure(figsize = (15,12))

# counting above
for i, count in enumerate(category_counts):
    plt.text(i, count + 0.1, str(count), ha='center', va='bottom',fontsize=20)

plt.bar(category_counts.index, category_counts.values)
plt.title('Youtube Contents counting per Category',fontsize=30)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df['date'] = pd.to_datetime(df['date'])

df['month'] = df['date'].dt.to_period('M')

uploads_by_month = df.groupby('month').size().reset_index(name='upload_count')

plt.figure(figsize=(10, 6))
sns.barplot(x='month', y='upload_count', data=uploads_by_month, color = 'skyblue')
plt.title('Monthly Upload Count')

plt.xlabel('Month')
plt.yticks([0,5,10,15,20])
plt.ylabel('Upload Count')
plt.show()
```
