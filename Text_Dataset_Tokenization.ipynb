### kkma

```python
from konlpy.tag import Kkma
kkma = Kkma() # 형태소 분석기 객체생성
```

```python
print(kkma.nouns(u"질문이나 건의사항은 깃헙 이슈 트래커에 남겨주세요")) # 명사
print(kkma.morphs(u"질문이나 건의사항은 깃헙 이슈 트래커에 남겨주세요")) # 다른 형태소도 추출
print(kkma.pos(u"질문이나 건의사항은 깃헙 이슈 트래커에 남겨주세요")) # mops에 품사정보가 추가됨 nng 일반명사, jc 조사 등
```

### Hannanum

```python
from konlpy.tag import Hannanum
hannanum = Hannanum() # 객체생성
```

```python
print(hannanum.morphs("질문이나 건의사항은 깃헙 이슈 트래커에 남겨주세요"))
print(hannanum.nouns("질문이나 건의사항은 깃헙 이슈 트래커에 남겨주세요"))
print(hannanum.pos("질문이나 건의사항은 깃헙 이슈 트래커에 남겨주세요"))
```

### klt2023

```python
from konlp.kma.klt2023 import klt2023
k = klt2023()
```

```python
simple_txt = "내 눈을 본다면 밤하늘의 별이 되는 기분을 느낄 수 있을 거야"

print(k.nouns(simple_txt))
print(k.pos(simple_txt))
print(k.morphs(simple_txt))
```
